```python
import datetime
birthday = input("Введите вашу дату рождения:")
Day, Month, Year = birthday.split('.')
Day = int(Day)
Month = int(Month)
Year = int(Year)
birthday_2 = datetime.datetime(Year, Month, Day)
today = datetime.datetime.today()
def calc_age(birthday_2):
    return today.year - birthday_2.year - ((today.month, today.day)\
                                          < (birthday_2.month, birthday_2.day))
def calc_dates(birthday_2, today):
    if today.day <= birthday_2.day and today.month <= birthday_2.month:
        delta1 = datetime.datetime(today.year, birthday_2.month, birthday_2.day)
        days = (delta1 - today).days
        return days
    else:
        delta2 = datetime.datetime(today.year + 1, birthday_2.month, birthday_2.day)
        days = (delta2 - today).days
        return days
delta = calc_age(birthday_2)
delta_2 = calc_dates(birthday_2, today)
print(f'Ваш возраст: {delta}')
print(f'Дней до Дня Рождения: {delta_2}')
```

```python
import random 
num = random.randint(1, 10) 
for t in range(1, 10): 
    quess = input('"""Загадано случайное число от 1 до 10. У вас 3 попытки, чтобы его угадать"""') 
    print(f'"""Попытка № {t}: """')
    if quess == num: 
        print(f'"""Ура, вы угадали! Это была цифра: {quess}') 
        break 
    elif quess.isdigit() == False:
        print(f'"""Нужно ввести число, а {quess} - не число"""')
    elif quess != num: 
        print(f'"""Вы не угадали!"""') 
    if t == 3:
        print(f'"""Было загадано число: {num}"""')
        break
```
